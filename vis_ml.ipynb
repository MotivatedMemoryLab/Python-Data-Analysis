# Introduction to `matplotlib` and `sklearn`

 `matplotlib` is the most commonly used visualization in Python, and `sklearn` is a powerful light-weight machine learning library which is super easy to use.

## `Matplotlib`

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# simple line graph
np.random.seed(0)
# random numbers
X = np.arange(10)
Y = np.random.rand(10)
# plot
plt.plot(X, Y)
```

```python
# scatter plots
plt.scatter(X, Y)
```

```python
# multiple lines in the same plot
Y = np.random.rand(10)
Z = np.random.rand(10)
plt.plot(X, Y, X, Z)
```

```python
# manipulate style
plt.plot(X, Y, 'r--', X, Z, 'b.-')
```
